First import the necessary packages:

```python
import numpy as np
import matplotlib.pylab as plt
```

Next we define the main Brownian motion simulation function:

```python
def simulate_Brownian_Motion(paths, steps, T):
    deltaT = T/steps
    t = np.linspace(0, T, steps+1)
    X = np.c_[np.zeros((paths, 1)),
              np.random.randn(paths, steps)]
    return t, np.cumsum(np.sqrt(deltaT) * X, axis=1)
```

Plot the simulated Brownian motion paths:

```python
plt.figure()
t, x = simulate_Brownian_Motion(5000, 100, 10.0)
plt.plot(t, x.T)
plt.show()
```

Brownian motion's distribution is given by:
$$W_t \sim N(0, t)$$
We can test this against our simulation.

```python
plt.figure()
plt.plot(t, np.mean(x, axis=0))
plt.xlabel('Time')
plt.ylabel(r'$\mathbb{E}[W_t]$')
plt.show()
```

The figure above shows that the mean of $W_t$ hovers around the value of 0. Next we plot the variance of $W_t$ over time, and show that variance increases linearly over time:

```python
plt.figure()
plt.plot(t, np.var(x, axis=0))
plt.xlabel('Time')
plt.ylabel(r'$V[W_t]$')
plt.show()
```
