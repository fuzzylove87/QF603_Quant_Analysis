```python
x=[1, 2, 3, 4]
#Q1: Compare x>1 element-by-element

#Q2: Is there any element of x greater than 1?

#Q3: Are all elements of x greater than 1?

#Q4: Compute x+1 element-by-element

#Q5: Compute x**2 element-by-element

#Q6: Test x in range(5) element-by-element

#Q8: Are all elements of x in range(5)
```

````python
x=[1, 2, 3, 4]
#Q1: Compare x>1 element-by-element
print([i>1 for i in x])
#Q2: Is there any element of x greater than 1?
print(any([i>1 for i in x]))
for s in x:
    if s>1:
        r=True
        break
else:
    r=False
print(r)
#ì¤‘ê°„ì— ì–´ë–¤ 1ì´ìƒì˜ ê°’ì´ë¼ë„ ë°œê²¬í•˜ë©´ ë°”ë¡œ breakí•˜ê³  êµ¬ë¬¸ì„ ë‚˜ì™€ë²„ë¦¼. ëê¹Œì§€ 1ì´ìƒì˜ ê°’ì´ ë‚˜ì˜¤ì§€ ì•Šì•˜ì„ ê²½ìš°ì—ë§Œ r=Falseê°€ ì ìš©ëœë‹¤.
#Q3: Are all elements of x greater than 1?
print(all([i>1 for i in x]))
#Q4: Compute x+1 element-by-element
print([i+1 for i in x])
#Q5: Compute x**2 element-by-element
print([i**2 for i in x])
#Q6: Test x in range(5) element-by-element
print([i in range(5) for i in x])
#Q7: Is there any element of x in range(5)
print(any([i in range(5) for i in x]))
#Q8: Are all elements of x in range(5)
print(all([i in range(5) for i in x]))
````

## Option Pricing (Binomial Tree)

#### Algorithm

Given $S$, $K$, $r$, $q$, $t$, $T$, $\sigma$ and $N$, (1 day = 1/365 years),

1. $\displaystyle \Delta t=\frac{T-t}{N}, u=e^{\sigma \sqrt{\Delta t}}, d=\frac{1}{u}, p=\frac{e^{(r-q)\Delta t}-d}{u-d}$
2. $\displaystyle f_{N,j} = \max (0, S_0 u^j d^{N-j}-K) \quad \mbox{(Call)}, \qquad f_{N,j} = \max (0, K-S_0 u^j d^{N-j}) \quad \mbox{(Put)},$ for $j=0, 1, ..., N$
3. $\displaystyle f_{i,j} = e^{-r\Delta t} [p\cdot f_{i+1,j+1} + (1-p)\cdot f_{i+1,j}] $ for $i=N-1, N-2, ..., 1, 0$; $j=0, 1, ..., i$
4. Option (call/put) price: $\displaystyle V=f_{0,0}$

(Let $\tau$ denote $T-t$.)

````python
# Binomial Tree Option Pricing (nested list with for loop)
import math
# T-t -> tau
def V_binomial_tree_py(S,K,r,q,tau,sigma,N=100):
   # 1.
   deltaT=tau/N
   u=math.exp(sigma*math.sqrt(deltaT))
   d=1/u
   p=(math.exp((r-q)*deltaT)-d)/(u-d)
   # 2.
   fc=[[0.0 for j in range(i+1)] for i in range(N+1)]
   fp=[[0.0 for j in range(i+1)] for i in range(N+1)]
   for j in range(N+1):
      fc[N][j]=max(0, S*(u**j)*(d**(N-j))-K)
      fp[N][j]=max(0, K-S*(u**j)*(d**(N-j)))
   # 3.
   p1=1-p
   ert=math.exp(-r*deltaT)
   for i in range(N-1,0-1,-1):
      for j in range(i+1):
         fc[i][j]=ert*(p*fc[i+1][j+1]+p1*fc[i+1][j])
         fp[i][j]=ert*(p*fp[i+1][j+1]+p1*fp[i+1][j])    
   # 4.
   c=fc[0][0]
   p=fp[0][0]
   return (c, p)

if __name__=='__main__':
   S=50.0; K=50.0; tau=183/365 
   sigma=0.4; r=0.04; q=0.01
   print('Call: {0[0]}, Put: {0[1]}'.format(
                V_binomial_tree_py(S,K,r,q,tau,sigma)))
````
