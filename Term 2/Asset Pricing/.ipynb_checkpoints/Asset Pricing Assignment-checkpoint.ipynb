```python
V = np.matrix(df1.cov())
R = np.matrix(df1.mean())
R = R.reshape(R.size, 1)
```

```python
rf = 0.13
```

```python
I = np.ones((len(R),1))
SD = np.sqrt(np.diag(V))
```

```python
Delta = I.T*pinv(V)*I 
Alpha = R.T*pinv(V)*I
Zeta = R.T*pinv(V)*R
D = Zeta*Delta-Alpha**2 #Denominator of Lagrange multipliers
```

```python
mu = np.arange(-max(R),max(R)*5,max(R)/100)
```

```python
minvar = (Zeta-2*Alpha*mu+(Delta*mu**2))/D;
minstd = np.sqrt(minvar)[0]
minstd = np.squeeze(np.asarray(minstd))
plt.plot(minstd,mu,SD, R, '*')
plt.yticks(np.arange(0, 3))
plt.title('Efficient frontier with Individual Securities',fontsize=20)
plt.ylabel('Expected Return (%)', fontsize=14)
plt.xlabel('Standard Deviation (%)', fontsize=14)
plt.show()
```
